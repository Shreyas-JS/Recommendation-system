```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
books= pd.read_csv('C:/Users/shrey/Downloads/Assignment/10.Recommendation system/book.csv')
books.head()
```

```python
books.drop(['Unnamed: 0'], axis=1, inplace=True)
books.head()
```

```python
books1 = books.rename({'User.ID': 'UserID','Book.Title':'Book_Title','Book.Rating':'Book_Rating'}, axis=1)
books1
```

```python
#number of unique users in the dataset
len(books1.UserID.unique())
```

```python
len(books1.Book_Title.unique())
```

```python
# drop all duplicte values 
books1.drop_duplicates(subset ="Book_Title",
                     keep = False, inplace = True)
books1
```

```python
user_books_df = books1.pivot(index='UserID',
                                 columns='Book_Title',
                                 values='Book_Rating').reset_index(drop=True)
user_books_df
```

```python
user_books_df.index = books1.UserID.unique()
user_books_df
```

```python
#Impute those NaNs with 0 values
user_books_df.fillna(0, inplace=True)
user_books_df
```

```python
#Calculating Cosine Similarity between Users
user_sim = 1 - pairwise_distances( user_books_df.values,metric='cosine')
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = books1.UserID.unique()
user_sim_df.columns = books1.UserID.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
